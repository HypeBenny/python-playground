```python
a = np.arange(1,7)
a
```

```python
b = a * 2
b
```

```python
plt.subplot(1,3,1)
plt.plot(a,b, c='red')
plt.subplot(1,3,2)
plt.plot(a,b)
plt.subplot(1,3,3)
plt.plot(a,b, 'green')
```

```python
plt.plot(a,b)
plt.plot(b,a)
plt.legend(['ab', 'ba'], loc=2) #if u chose 0, it chooses the right location for u.
plt.text(6,6, 'heya')           #use this to add a text to the fxn. use 6,6 to specify position
plt.grid(True, color='y')
plt.savefig('ben')       #how to save a chart.
```

```python
plt.plot(a,b)
plt.plot(b,a)
plt.legend(['ab', 'ba'], loc=2) #if u chose 0, it chooses the right location for u.
plt.text(5,7, '>>> heya')         #use this to add a text to the fxn. use 6,6 to specify position based on the axis number labels.
plt.xlim([0,6])      #used to set limit of numbers to display in x n y axis
plt.ylim([0,8])
```

```python
names =['adjoa', 'ama', 'kofi', 'ken']
scores = [200, 140, 300, 180]
attention =[0, 0, 0.1, 0]   #attention here is used to create emphasis on a particular slice in the pie chart. we wanna create emphasis on the 3rd itwm in the list so we set all to 0, then set the 3rd element to 0.1
colors= ['pink', 'purple', 'yellow', 'green']  #not necesary u can remove n plt will give default color.
plt.pie(scores, labels=names, explode=attention, autopct='%1.1f%%', colors=colors, startangle=90, shadow=True, wedgeprops={'edgecolor':'white'});  #here u pass a new parameter called explode to help create the emphasis
#shadow para is used to create that 3d effect
#start angle used to rotate it. indicate the starting angle.
#autopct para used to disolay %age inside.
#use pie chart if few variables are being compared.
```
