```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
```

```python
yields = [0.3, 0.5, 0.6, 0.7, 0.75, 0.82]
years = [2012, 2013, 2014, 2015, 2016, 2017]
```

```python

plt.plot(years, yields, marker='o', color= 'green', linestyle= '--', linewidth=4, markersize=10, markerfacecolor='yellow', markeredgecolor='black', markeredgewidth=2, alpha=.7)
plt.xlabel('year')
plt.ylabel('yields per year');

```

```python
plt.figure(figsize=(4,2))
plt.plot(years, yields)
sns.set_style('whitegrid')

```
