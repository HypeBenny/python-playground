```python
sampledata[0,0] == sampledata[0,1]
```

```python
sampledata  #3 columns here reps amount, units and totl cost.
```

```python
#profit = amount * units
profit = sampledata[:,0] * sampledata[:,1]
profit
```

```python
profit.sum()
```

```python
profit.mean()
```

```python
#now we add profit col back to sampledata.
profit_col = profit.reshape(300,1)
profit_col
```

```python
newdata = np.concatenate((sampledata, profit_col), axis=1)
newdata
```

```python
#we can find total cost sum.
sampledata[:,2].sum()
```

```python
#we can find average of total cost.
sampledata[:,2].mean()
```

```python
#we create a new column called tripled units which is units * 3.
triple_units = sampledata[:,1] * 3
triple_units_col = triple_units.reshape(300,1)
```

```python
#we add it to newdata
dataset = np.concatenate((newdata, triple_units_col), axis=1)
```

```python
#now we save all back to the file.
np.savetxt('C:\\Users\Acer\Downloads\practice data.txt', dataset, fmt='%.f', delimiter=',          ', comments='', header='AMOUNT UNITS TOTAL_COST PROFIT TRIPLE_UNITS')
```
